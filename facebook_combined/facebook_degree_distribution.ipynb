```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist("facebook_combined.txt.gz", create_using=nx.Graph(), nodetype=int)
```

```python
list_of_degrees = nx.degree_histogram(G)
```

```python


plt.figure(figsize=(12, 8))
plt.bar(range(len(list_of_degrees)), list_of_degrees)
#plt.hist(list_of_degrees, bins=100)
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.yscale("log")
plt.xscale("log")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
list_of_degrees = [x/4039 for x in list_of_degrees]
plt.plot(range(len(list_of_degrees)), list_of_degrees)
#plt.hist(list_of_degrees, bins=100)
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.yscale("log")
plt.xscale("log")
plt.show()
```
